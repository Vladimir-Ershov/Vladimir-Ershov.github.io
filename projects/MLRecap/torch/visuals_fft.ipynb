```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 1000)

# Define different frequencies for complex exponentials
frequencies = [0.5, 1.0, 1.5, 2.0]

# Create figure with multiple subplots
fig = plt.figure(figsize=(16, 12))

# Plot 1: Real and Imaginary parts separately
plt.subplot(2, 3, (1, 2))
for freq in frequencies:
    # Complex exponential: e^(i*freq*x)
    complex_exp = np.exp(1j * freq * x)
    
    plt.plot(x, np.real(complex_exp), label=f'Real(e^(i*{freq}*x))', linewidth=2)
    plt.plot(x, np.imag(complex_exp), '--', label=f'Imag(e^(i*{freq}*x))', linewidth=2)

plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Complex Exponentials: Real and Imaginary Parts')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-10, 10)

# Plot 2: Magnitude
plt.subplot(2, 3, 3)
for freq in frequencies:
    complex_exp = np.exp(1j * freq * x)
    magnitude = np.abs(complex_exp)
    plt.plot(x, magnitude, label=f'|e^(i*{freq}*x)|', linewidth=2)

plt.xlabel('x')
plt.ylabel('Magnitude')
plt.title('Magnitude of Complex Exponentials')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-10, 10)

# Plot 3: Phase
plt.subplot(2, 3, 4)
for freq in frequencies:
    complex_exp = np.exp(1j * freq * x)
    phase = np.angle(complex_exp)
    plt.plot(x, phase, label=f'e^(i*{freq}*x)', linewidth=2)

plt.xlabel('x')
plt.ylabel('Phase (radians)')
plt.title('Phase of Complex Exponentials')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-10, 10)

# Plot 4: 3D visualization of one complex exponential in complex plane
ax = plt.subplot(2, 3, 5, projection='3d')
freq = 1.0
complex_exp = np.exp(1j * freq * x)

ax.plot(x, np.real(complex_exp), np.imag(complex_exp), linewidth=2, color='blue')
ax.set_xlabel('x (parameter)')
ax.set_ylabel('Real part')
ax.set_zlabel('Imaginary part')
ax.set_title(f'3D: e^(i*{freq}*x) in Complex Space')

# Plot 5: Complex plane trajectory (unit circle)
plt.subplot(2, 3, 6)
freq = 1.0
# Just show one complete revolution for clarity
x_circle = np.linspace(0, 2*np.pi, 100)
complex_exp = np.exp(1j * freq * x_circle)

plt.plot(np.real(complex_exp), np.imag(complex_exp), 'b-', linewidth=2, label=f'e^(i*{freq}*x)')
plt.plot(np.real(complex_exp[0]), np.imag(complex_exp[0]), 'go', markersize=8, label='Start')
plt.plot(np.real(complex_exp[-1]), np.imag(complex_exp[-1]), 'ro', markersize=8, label='End')

# Add arrows to show direction
for i in range(0, len(complex_exp), 20):
    dx = np.real(complex_exp[min(i+5, len(complex_exp)-1)]) - np.real(complex_exp[i])
    dy = np.imag(complex_exp[min(i+5, len(complex_exp)-1)]) - np.imag(complex_exp[i])
    plt.arrow(np.real(complex_exp[i]), np.imag(complex_exp[i]), dx, dy, 
              head_width=0.05, head_length=0.05, fc='red', ec='red', alpha=0.7)

plt.xlabel('Real part')
plt.ylabel('Imaginary part')
plt.title('Complex Exponential in Complex Plane')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')

plt.tight_layout()
plt.show()

# Print some key properties
print("Key Properties of Complex Exponentials e^(i**x):")
print("=" * 50)
print("1. Magnitude is always 1: |e^(i**x)| = 1")
print("2. Real part: cos(*x)")
print("3. Imaginary part: sin(*x)")
print("4. Euler's formula: e^(i**x) = cos(*x) + i*sin(*x)")
print("5. They trace out circles in the complex plane")
print("6. Higher frequency  = faster rotation around unit circle")

# Show the connection to FFT basis functions
print("\nConnection to FFT:")
print("=" * 20)
print("FFT uses basis functions: e^(-i*2*k*n/N)")
print("- k: frequency bin (0, 1, 2, ...)")  
print("- n: time sample (0, 1, 2, ..., N-1)")
print("- N: window size")
print("Each k gives a different rotation speed around the unit circle")
```
